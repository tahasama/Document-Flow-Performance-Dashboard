# Document Flow Performance Analysis

This report evaluates how documents move from **planned submission → review → approval**.

**Audience**: Management, discipline leads, project engineers

The objective is to:

- Identify where delays are occurring  
- Distinguish between **supplier-driven** and **review-driven** delays  
- Highlight risk areas (discipline, backlog, SLA breaches)  
- Support **targeted operational decisions**, not generic conclusions  

## Quick definitions

- **Supplier submission delay:** a document was delivered later than the planned submission date. This can mean the supplier missed the plan and the team did not request the document at the time OR schedule moved for some reason — both cause the same delivery impact.
- **Review delay:** the document was not approved by the reviewers (internal or external) before the due date — an accountability/capacity issue.
- **Median cycle time:** the middle value of end-to-end delivery time (less sensitive to a few extreme outliers).
- **Backlog pressure:** percent of documents past their due date and still not approved.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,6)
```

```python
# LOAD
supplier_docs_raw = pd.read_excel("C:/Users/Taha/Desktop/Dashboard/tablesForTesting/ExportSupplierDocuments.xlsx")
workflow_raw = pd.read_excel("C:/Users/Taha/Desktop/Dashboard/tablesForTesting/ExportWorkflow.xlsx")
```

```python
# CLEAN
def clean_header(df):
    df = df.iloc[7:].copy()
    df.columns = df.iloc[0]
    df = df[1:]
    df.columns = df.columns.str.strip()
    df.reset_index(drop=True, inplace=True)
    return df

supplier_docs = clean_header(supplier_docs_raw)
workflow = clean_header(workflow_raw)
```

```python
# FORMAT
supplier_docs["Document No"] = supplier_docs["Document No"].astype(str)
workflow["Document No."] = workflow["Document No."].astype(str)

supplier_docs["Planned Submission Date"] = pd.to_datetime(
    supplier_docs["Planned Submission Date"], errors="coerce"
)

workflow["Date In"] = pd.to_datetime(workflow["Date In"], errors="coerce")
workflow["Date Completed"] = pd.to_datetime(workflow["Date Completed"], errors="coerce")
workflow["Date Due"] = pd.to_datetime(workflow["Date Due"], errors="coerce")
```

```python
# BUILD DATASET → df
workflow = workflow.rename(columns={
    "Document No.": "Document No",
    "Assigned To": "Assigned_To"
})

agg = workflow.groupby("Document No").agg({
    "Date In": "min",
    "Date Completed": "max",
    "Date Due": "max",
    "Assigned_To": "first"
}).reset_index()

agg = agg.rename(columns={
    "Date In": "FirstDateIn",
    "Date Completed": "ApprovedDate",
    "Date Due": "DueDate"
})

df = supplier_docs.merge(agg, on="Document No", how="left")
```

## Performance Overview

This section answers a simple question:

**Are we delivering documents on time across submission and review?**

### KPI Definitions

- **Supplier On-Time %**  
  % of documents submitted on or before the planned date  

- **Reviewer On-Time % (SLA)**  
  % of documents approved before their due date  
  *(SLA = agreed review deadline)*  

- **Median Cycle Time**  
  Typical number of days from planned submission to approval  
  *(Median avoids distortion from extreme delays)*  

- **Backlog Pressure**  
  % of documents currently overdue and not yet completed  

```python
# KPI
df["SupplierOnTime"] = df["FirstDateIn"] <= df["Planned Submission Date"]
supplier_on_time = df["SupplierOnTime"].mean() * 100

df["ReviewerOnTime"] = df["ApprovedDate"] <= df["DueDate"]
review_on_time = df["ReviewerOnTime"].mean() * 100

df["CycleTime"] = (df["ApprovedDate"] - df["Planned Submission Date"]).dt.days
median_cycle = df["CycleTime"].median()

df["Overdue"] = (df["DueDate"] < pd.Timestamp.today()) & (
    df["ApprovedDate"].isna() | (df["ApprovedDate"] > df["DueDate"])
)
backlog = df["Overdue"].mean() * 100

print(f"Supplier On-Time: {supplier_on_time:.1f}%")
print(f"Review On-Time (SLA): {review_on_time:.1f}%")
print(f"Median Cycle Time: {median_cycle:.0f} days")
print(f"Backlog Pressure: {backlog:.1f}%")
```

## Where Is Time Being Spent?

Total process time is split into:

- **Supplier Delay** → delay before submission  
- **Review Duration** → time inside review cycle  

This identifies the dominant source of delay.

```python
valid = df.dropna(subset=[
    "Planned Submission Date",
    "FirstDateIn",
    "ApprovedDate"
]).copy()

valid["SupplierDelay"] = (
    valid["FirstDateIn"] - valid["Planned Submission Date"]
).dt.days.clip(lower=0)

valid["ReviewDuration"] = (
    valid["ApprovedDate"] - valid["FirstDateIn"]
).dt.days.clip(lower=0)

valid["Total"] = valid["SupplierDelay"] + valid["ReviewDuration"]
valid = valid[valid["Total"] > 0]

supplier_pct = valid["SupplierDelay"].sum() / valid["Total"].sum() * 100
review_pct = valid["ReviewDuration"].sum() / valid["Total"].sum() * 100

print(f"Supplier Contribution: {supplier_pct:.1f}%")
print(f"Review Contribution: {review_pct:.1f}%")
```

```python
plt.bar(["Supplier", "Review"], [supplier_pct, review_pct])
plt.title("Contribution to Total Cycle Time")
plt.ylabel("% of Total Duration")
plt.show()
```

## Flow Stability (Submission vs Approval)

This compares:

- Documents entering the system  
- Documents being completed  

If submissions exceed approvals → backlog grows  
If approvals exceed submissions → backlog reduces  

```python
flow = df.dropna(subset=["FirstDateIn", "ApprovedDate"]).copy()

flow["Week_Sub"] = flow["FirstDateIn"].dt.to_period("W").dt.start_time
flow["Week_App"] = flow["ApprovedDate"].dt.to_period("W").dt.start_time

submitted = flow.groupby("Week_Sub").size()
approved = flow.groupby("Week_App").size()

trend = pd.concat([submitted, approved], axis=1, sort=True).fillna(0)
trend.columns = ["Submitted", "Approved"]

trend["Backlog"] = (trend["Submitted"] - trend["Approved"]).cumsum()

trend.plot()
plt.title("Submission vs Approval Trend")
plt.show()
```

## Discipline Risk (SLA Breach)

This shows which disciplines are driving delays.

Focus is on:
- Late approvals
- Frequency of SLA breaches  

```python
df_sla = df.dropna(subset=["DueDate", "ApprovedDate"]).copy()

df_sla["Delay"] = (df_sla["ApprovedDate"] - df_sla["DueDate"]).dt.days
df_sla["Delay"] = df_sla["Delay"].clip(lower=0)

risk = df_sla.groupby("Select List 1").agg(
    Total=("Document No", "count"),
    Late=("Delay", lambda x: (x > 0).sum())
)

risk["Breach_%"] = risk["Late"] / risk["Total"] * 100
risk = risk.sort_values("Breach_%", ascending=False)

risk.head()
```

```python
# Heatmap
pivot = pd.pivot_table(
    df_sla,
    values="Document No",
    index="Select List 1",
    columns=pd.cut(df_sla["Delay"], bins=[0,5,9,100], labels=["1-5","6-9","10+"]),
    aggfunc="count",
    fill_value=0
)

sns.heatmap(pivot, annot=True, fmt="g")
plt.title("Delay Distribution by Discipline")
plt.show()
```

## Reviewer Performance

This section evaluates:

- Average review duration per reviewer  
- SLA compliance  

This highlights accountability and workload distribution  

```python
df_rev = df.dropna(subset=["FirstDateIn", "ApprovedDate", "Assigned_To"]).copy()

df_rev["ReviewDays"] = (df_rev["ApprovedDate"] - df_rev["FirstDateIn"]).dt.days

review_perf = df_rev.groupby("Assigned_To").agg(
    Avg_Days=("ReviewDays", "mean"),
    Count=("Document No", "count")
).sort_values("Avg_Days", ascending=False)

review_perf.head()
```

```python
sns.barplot(
    data=review_perf.reset_index(),
    x="Avg_Days",
    y="Assigned_To"
)
plt.title("Average Review Duration by Reviewer")
plt.show()
```

## Distribution of Review Durations

This shows how review time is spread:

- Fast reviews  
- Typical range  
- Extreme delays  

Used to detect inconsistency and outliers  

```python
dist = df_rev["ReviewDays"].dropna()

bins = [0,5,10,20,100]
labels = ["0-5","6-10","11-20","20+"]

bucket = pd.cut(dist, bins=bins, labels=labels, right=False)
counts = bucket.value_counts().sort_index()

sns.barplot(x=counts.index, y=counts.values)
plt.title("Review Duration Distribution")
plt.show()

print((counts / counts.sum() * 100).round(1))
```

## Key Observations (Generated from Data)

- If review contribution is higher → issue is internal process or capacity  
- If supplier contribution is higher → issue is planning / follow-up  

- High backlog + low approval rate → system overload  
- High SLA breach in specific discipline → localized issue  

These results should be used to **trigger focused actions**, not generic improvements  

```python
print("=== ACTION TRIGGERS ===\n")

if review_pct > supplier_pct:
    print("- Review process is the dominant delay → check reviewer capacity & prioritization")

if supplier_pct > review_pct:
    print("- Supplier delay is dominant → enforce submission planning & follow-up discipline")

if backlog > 20:
    print("- High backlog → immediate clearing plan required")

top_discipline = risk.index[0]
print(f"- Highest risk discipline: {top_discipline} → requires focused intervention")
```
