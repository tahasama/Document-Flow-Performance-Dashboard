## Notebook Name: Document_Flow_Analysis.ipynb
Purpose: Compute KPIs, visualize delays, show root causes (supplier vs reviewer), and generate actionable insights.

Audience: Management, discipline leads, project engineers

### Environment Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: for nicer charts in notebook
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,6)
```

### Load Dataset

```python
# Example: CSV exports from EDMS / workflow
supplier_docs_raw = pd.read_excel("C:/Users/Taha/Desktop/Dashboard/tablesForTesting/ExportSupplierDocuments.xlsx")
workflow_raw = pd.read_excel("C:/Users/Taha/Desktop/Dashboard/tablesForTesting/ExportWorkflow.xlsx")
```

### Clean Structure & Fix Headers 
#### Remove first 8 rows and set row 9 as header in both files

The raw file contains metadata in the first rows.

Actions:
1. Remove the first 9 rows
2. Use the new first row as column headers
3. Reset index for clean structure

This transforms the dataset into a usable tabular format.

```python
supplier_docs = supplier_docs_raw.iloc[7:].copy()
supplier_docs.columns = supplier_docs.iloc[0]  # set header
supplier_docs = supplier_docs[1:]  # remove header row from data

# Reset index
supplier_docs.reset_index(drop=True, inplace=True)
```

```python
workflow = workflow_raw.iloc[7:].copy()
workflow.columns = workflow.iloc[0]
workflow = workflow[1:]

# Reset index
workflow.reset_index(drop=True, inplace=True)
```

### Clean columns name

```python
supplier_docs.columns = supplier_docs.columns.str.strip()
workflow.columns = workflow.columns.str.strip()
```

## Standardize Key Fields

We align both tables on a common document key:

- Document No → unique identifier across both datasets

We also ensure all date fields are in proper datetime format to avoid calculation errors.

```python
# Ensure key column is string (safe for joins/grouping)
supplier_docs["Document No"] = supplier_docs["Document No"].astype(str)
workflow["Document No."] = workflow["Document No."].astype(str)

# Convert date columns (adjust names if needed)
supplier_docs["Planned Submission Date"] = pd.to_datetime(
    supplier_docs["Planned Submission Date"], errors="coerce"
)

workflow["Date In"] = pd.to_datetime(workflow["Date In"], errors="coerce")
workflow["Date Completed"] = pd.to_datetime(workflow["Date Completed"], errors="coerce")
workflow["Date Due"] = pd.to_datetime(workflow["Date Due"], errors="coerce")
```

### Build Document-Level Dates

Workflow contains multiple rows per document.

We aggregate it to document level:

- FirstDateIn → first time document entered workflow (submission proxy)
- ApprovedDate → when review completed
- DueDate → review deadline

This creates a clean dataset for KPI calculations.

```python
# 1️⃣ Ensure workflow column is correctly named
workflow = workflow.rename(columns={"Document No.": "Document No", "Assigned To": "Assigned_To"})

# Function to pick first assigned reviewer
def first_assigned(x):
    return x.iloc[0]

# 2️⃣ Aggregate workflow to document level
doc_dates = workflow.groupby("Document No").agg({
    "Date In": "min",          # First submission
    "Date Completed": "max",   # Approved date
    "Date Due": "max",          # Due date
    "Assigned_To": first_assigned 
}).reset_index()

# 3️⃣ Rename columns: only the date columns, and fix key if needed
doc_dates = doc_dates.rename(columns={
    "Document No.": "Document No",  # fix potential leftover dot
    "Date In": "FirstDateIn",
    "Date Completed": "ApprovedDate",
    "Date Due": "DueDate"
})
```

## Merge Supplier & Workflow Data

We combine:

- Planning data (supplier_docs)
- Actual workflow data (doc_dates)

Result: one row per document with full lifecycle visibility.

```python
# 4️⃣ Verify doc_dates columns
print("Doc Dates Columns after rename:", doc_dates.columns.tolist())
# Expected output: ['Document No', 'FirstDateIn', 'ApprovedDate', 'DueDate']

# 5️⃣ Merge with supplier_docs
df = supplier_docs.merge(doc_dates, on="Document No", how="left")
```

# KPI 1: Supplier On-Time %

Definition:  
A supplier is on time if **Actual Submission (FirstDateIn) ≤ Planned Submission Date**.

We calculate the percentage of documents submitted on time.

```python
# Supplier on-time flag
df["SupplierOnTime"] = df["FirstDateIn"] <= df["Planned Submission Date"]

# KPI calculation
supplier_on_time_pct = df["SupplierOnTime"].mean() * 100

print(f"Supplier On-Time %: {supplier_on_time_pct:.2f}%")
```

## KPI 2: Average Supplier Delay

Definition:  
Average delay in days for **late submissions only**:

Delay = FirstDateIn − Planned Submission Date

```python
# Only late documents
late_docs = df[df["FirstDateIn"] > df["Planned Submission Date"]]

# Average delay
avg_supplier_delay = (late_docs["FirstDateIn"] - late_docs["Planned Submission Date"]).dt.days.mean()

print(f"Average Supplier Delay (days): {avg_supplier_delay:.2f}")
```

## KPI 3: Reviewer On-Time %

Definition:  
A review is on time if **ApprovedDate ≤ DueDate**.

This shows internal performance.

```python
df["ReviewerOnTime"] = df["ApprovedDate"] <= df["DueDate"]

reviewer_on_time_pct = df["ReviewerOnTime"].mean() * 100

print(f"Reviewer On-Time %: {reviewer_on_time_pct:.2f}%")
```

## KPI 4: Median Total Cycle Time

Definition:  
Total Cycle Time = ApprovedDate − Planned Submission Date  
Median is used to reduce distortion from extreme outliers.

```python
df["TotalCycleTime"] = (df["ApprovedDate"] - df["Planned Submission Date"]).dt.days

median_total_cycle = df["TotalCycleTime"].median()

print(f"Median Total Cycle Time (days): {median_total_cycle}")
```

## KPI 5: Backlog Pressure

Definition:  
Percentage of documents currently **past due date** and not yet approved.

```python
df["Overdue"] = (df["DueDate"] < pd.Timestamp.today()) & (df["ApprovedDate"].isna() | (df["ApprovedDate"] > df["DueDate"]))

backlog_pct = df["Overdue"].mean() * 100

print(f"Backlog Pressure (% overdue): {backlog_pct:.2f}%")
```

## Section A — Where Is Time Lost?

We compute three durations per document:

1. Supplier Delay: Planned Submission → First Submission  
2. Review Duration: First Submission → Approved Date  
3. Total Lead Time: Planned Submission → Approved Date  

We will show **average or median per document** in a clustered bar chart to reveal where delays occur.

Code — Compute Durations

```python
# Ensure datetime (safety)
df["Planned Submission Date"] = pd.to_datetime(df["Planned Submission Date"], errors="coerce")
df["FirstDateIn"] = pd.to_datetime(df["FirstDateIn"], errors="coerce")
df["ApprovedDate"] = pd.to_datetime(df["ApprovedDate"], errors="coerce")

# Compute Review Duration
df["Review_Duration_Days"] = (df["ApprovedDate"] - df["FirstDateIn"]).dt.days

# Keep valid rows
df_clean = df.dropna(subset=["FirstDateIn", "ApprovedDate", "Assigned_To", "Review_Duration_Days"])
df_clean = df_clean[df_clean["Review_Duration_Days"] >= 0].copy()

# Quick check
print(df_clean[["Document No", "Assigned_To", "Review_Duration_Days"]].head())

# Keep only valid rows
valid_df = df.dropna(subset=[
    "Planned Submission Date",
    "FirstDateIn",
    "ApprovedDate"
]).copy()

# Compute durations
valid_df["SupplierDelay"] = (valid_df["FirstDateIn"] - valid_df["Planned Submission Date"]).dt.days
valid_df["ReviewDuration"] = (valid_df["ApprovedDate"] - valid_df["FirstDateIn"]).dt.days
valid_df["TotalLeadTime"] = (valid_df["ApprovedDate"] - valid_df["Planned Submission Date"]).dt.days
```

## Stage Duration Breakdown

Median duration per stage is used to avoid distortion from outliers.

This reveals where most time is spent:
- Supplier
- Review
- Total process

```python
# Median values
stage_medians = valid_df[[
    "SupplierDelay",
    "ReviewDuration",
    "TotalLeadTime"
]].median()

# Plot (clean matplotlib)
stage_medians.plot(kind="bar")

plt.title("Stage Duration Breakdown (Median Days)")
plt.ylabel("Days")
plt.xlabel("Stage")
plt.xticks(rotation=0)
plt.show()
```

## Flow Contribution %

We measure how much each stage contributes to the total process time:

- Supplier Contribution % = Supplier Delay / Total Lead Time  
- Review Contribution % = Review Duration / Total Lead Time  

This identifies the dominant bottleneck in the process.

```python
# Avoid division by zero
valid_df = valid_df[valid_df["TotalLeadTime"] > 0]

# Compute contribution %
valid_df["SupplierContribution"] = valid_df["SupplierDelay"] / valid_df["TotalLeadTime"]
valid_df["ReviewContribution"] = valid_df["ReviewDuration"] / valid_df["TotalLeadTime"]

# Average contribution (as %)
supplier_pct = valid_df["SupplierContribution"].mean() * 100
review_pct = valid_df["ReviewContribution"].mean() * 100

print(f"Supplier Contribution: {supplier_pct:.1f}%")
print(f"Review Contribution: {review_pct:.1f}%")
```

## Executive Insight

We automatically generate a clear conclusion based on contribution:

- If review dominates → internal bottleneck  
- If supplier dominates → external bottleneck  

```python
if review_pct > supplier_pct:
    print(f"Review process is the main bottleneck, accounting for {review_pct:.0f}% of total cycle time.")
else:
    print(f"Supplier delay is the main bottleneck, accounting for {supplier_pct:.0f}% of total cycle time.")
```

```python
# Simple bar chart
contrib = pd.Series({
    "Supplier": supplier_pct,
    "Review": review_pct
})

contrib.plot(kind="bar")

plt.title("Flow Contribution (%)")
plt.ylabel("Percentage of Total Cycle Time")
plt.xticks(rotation=0)
plt.show()
```

## Section B — Flow Trend Over Time

We track weekly flow:

- Documents Submitted (FirstDateIn)
- Documents Approved (ApprovedDate)

Insight:

- If Submitted > Approved → backlog increases  
- If Approved > Submitted → backlog decreases  

This shows system stability over time.

```python
# ==============================
# SECTION B — WEEKLY FLOW DATA
# ==============================

# Ensure datetime (safe)
df["FirstDateIn"] = pd.to_datetime(df["FirstDateIn"], errors="coerce")
df["ApprovedDate"] = pd.to_datetime(df["ApprovedDate"], errors="coerce")

# Keep only valid rows
df_flow = df.dropna(subset=["FirstDateIn", "ApprovedDate"]).copy()

# Create week columns
df_flow["Submitted_Week"] = df_flow["FirstDateIn"].dt.to_period("W").dt.start_time
df_flow["Approved_Week"] = df_flow["ApprovedDate"].dt.to_period("W").dt.start_time

# Count per week
submitted = df_flow.groupby("Submitted_Week").size()
approved = df_flow.groupby("Approved_Week").size()

# Combine into one table
flow_weekly = pd.concat([submitted, approved], axis=1)
flow_weekly.columns = ["Submitted", "Approved"]

# Clean
flow_weekly = flow_weekly.fillna(0).sort_index().reset_index()
flow_weekly = flow_weekly.rename(columns={"index": "Week"})

# Preview
flow_weekly.head()
```

```python
# ==============================
# FLOW TREND CHART
# ==============================

plt.figure()

plt.plot(flow_weekly["Week"], flow_weekly["Submitted"], marker="o")
plt.plot(flow_weekly["Week"], flow_weekly["Approved"], marker="s")

plt.title("Submitted vs Approved Documents Over Time")
plt.xlabel("Week")
plt.ylabel("Documents")
plt.xticks(rotation=45)

plt.legend(["Submitted", "Approved"])

plt.tight_layout()
plt.show()
```

```python
# ==============================
# BACKLOG INDICATOR
# ==============================

flow_weekly["Backlog_Change"] = flow_weekly["Submitted"] - flow_weekly["Approved"]
flow_weekly["Backlog"] = flow_weekly["Backlog_Change"].cumsum()

flow_weekly.head()
```

```python
# ==============================
# BACKLOG TREND
# ==============================

plt.figure()

plt.plot(flow_weekly["Week"], flow_weekly["Backlog"], marker="o")

plt.title("Backlog Evolution")
plt.xlabel("Week")
plt.ylabel("Backlog Volume")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# ==============================
# SECTION C — SLA & Aging RISK (adapted)
# ==============================

# ==============================
# SECTION C — DATA PREPARATION
# ==============================

# Ensure datetime
df["DueDate"] = pd.to_datetime(df["DueDate"], errors="coerce")
df["ApprovedDate"] = pd.to_datetime(df["ApprovedDate"], errors="coerce")

# Keep valid rows
df_sla = df.dropna(subset=["DueDate", "ApprovedDate"]).copy()

# SLA compliance (1 = on time, 0 = late)
df_sla["OnTime"] = (df_sla["ApprovedDate"] <= df_sla["DueDate"]).astype(int)

# Ensure datetime
df_sla["DueDate"] = pd.to_datetime(df_sla["DueDate"], errors="coerce")
df_sla["ApprovedDate"] = pd.to_datetime(df_sla["ApprovedDate"], errors="coerce")

# Keep valid rows
df_sla = df_sla.dropna(subset=["DueDate", "ApprovedDate"]).copy()

# SLA compliance (1 = on time, 0 = late)
df_sla["OnTime"] = (df_sla["ApprovedDate"] <= df_sla["DueDate"]).astype(int)

# Delay in days (only late docs)
df_sla["Delay_Days"] = (df_sla["ApprovedDate"] - df_sla["DueDate"]).dt.days
df_sla["Delay_Days"] = df_sla["Delay_Days"].apply(lambda x: x if x > 0 else 0)

df_sla.head()
```

```python
# ==============================
# AGING BUCKETS
# ==============================

def aging_bucket(days):
    if days == 0:
        return "On Time"
    elif days <= 3:
        return "1-3 Days Late"
    elif days <= 7:
        return "4-7 Days Late"
    else:
        return "7+ Days Late"

df_sla["Aging_Bucket"] = df_sla["Delay_Days"].apply(aging_bucket)

df_sla[["Delay_Days", "Aging_Bucket"]].head()
```

```python
# ==============================
# HEATMAP TABLE
# ==============================

print(df_sla.columns.tolist())

aging_matrix = pd.pivot_table(
    df_sla,
    values="Document No",       
    index="Select List 1",        # discipline column
    columns="Aging_Bucket",
    aggfunc="count",
    fill_value=0
)

#aging_matrix
```

```python
# ==============================
# HEATMAP VISUAL
# ==============================

plt.figure()

sns.heatmap(aging_matrix, annot=True, fmt="g")

plt.title("Aging Distribution by Discipline")
plt.xlabel("Delay Category")
plt.ylabel("Discipline")

plt.tight_layout()
plt.show()
```

```python
# ==============================
# SLA BREACH %
# ==============================

sla_perf = (
    df_sla.groupby("Select List 1")
    .agg(
        Total=("Document No", "count"),
        OnTime=("OnTime", "sum")
    )
)

sla_perf["SLA_Compliance_%"] = sla_perf["OnTime"] / sla_perf["Total"] * 100
sla_perf["SLA_Breach_%"] = 100 - sla_perf["SLA_Compliance_%"]

sla_perf = sla_perf.sort_values("SLA_Breach_%", ascending=False)

sla_perf.head()
```

```python
# ==============================
# BREACH BAR CHART
# ==============================

plt.figure()

plt.bar(sla_perf.index, sla_perf["SLA_Breach_%"])

plt.title("SLA Breach % by Discipline")
plt.xlabel("Discipline")
plt.ylabel("Breach %")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## SECTION D — Reviewer Performance

```python
# Ensure datetime
df_clean["FirstDateIn"] = pd.to_datetime(df_clean["FirstDateIn"], errors="coerce")
df_clean["ApprovedDate"] = pd.to_datetime(df_clean["ApprovedDate"], errors="coerce")

# Compute review duration
df_clean["Review_Duration_Days"] = (df_clean["ApprovedDate"] - df_clean["FirstDateIn"]).dt.days

# Keep only valid durations
df_reviewer = df_clean.dropna(subset=["Review_Duration_Days", "Assigned_To"])
df_reviewer = df_reviewer[df_reviewer["Review_Duration_Days"] >= 0].copy()

# Aggregate by Reviewer
reviewer_perf = (
    df_reviewer.groupby("Assigned_To")
    .agg(
        Avg_Review_Days=("Review_Duration_Days", "mean"),
        Doc_Count=("Document No", "count")
    )
    .reset_index()
    .sort_values("Avg_Review_Days", ascending=False)
)

reviewer_perf.head()
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=reviewer_perf,
    x="Avg_Review_Days",
    y="Assigned_To",
    palette="coolwarm"
)

plt.title("Average Review Duration by Reviewer")
plt.xlabel("Average Review Duration (Days)")
plt.ylabel("Reviewer (Assigned To)")
plt.tight_layout()
plt.show()
```

```python
# SLA compliance: ApprovedDate <= DueDate
df_clean["OnTime"] = (df_clean["ApprovedDate"] <= df_clean["DueDate"]).astype(int)

sla_reviewer = df_clean.groupby("Assigned_To").agg(
    SLA_Compliance_Percent=("OnTime", lambda x: x.mean() * 100)
).sort_values("SLA_Compliance_Percent", ascending=True).reset_index()

# Bar chart
plt.figure(figsize=(10,6))
sns.barplot(
    data=sla_reviewer,
    x="SLA_Compliance_Percent",
    y="Assigned_To",
    palette="Reds_r"
)
plt.title("Reviewer SLA Compliance %")
plt.xlabel("SLA Compliance (%)")
plt.ylabel("Reviewer (Assigned To)")
plt.tight_layout()
plt.show()
```

## SECTION E — Distribution View

```python
# Use df_clean from Section D
# Ensure Review_Duration_Days is valid
distribution_df = df_clean[["Review_Duration_Days"]].copy()

# Quick stats check
print(distribution_df.describe())
```

```python
# Histogram bins (example: 0-5, 6-10, 11-20, 20+ days)
bins = [0, 5, 10, 20, distribution_df["Review_Duration_Days"].max()+1]
labels = ["0-5", "6-10", "11-20", "20+"]

distribution_df["Duration_Bucket"] = pd.cut(distribution_df["Review_Duration_Days"], bins=bins, labels=labels, right=False)

# Count per bucket
bucket_counts = distribution_df["Duration_Bucket"].value_counts().sort_index()
print(bucket_counts)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=bucket_counts.index, y=bucket_counts.values, palette="Blues_d")
plt.title("Distribution of Review Durations")
plt.xlabel("Review Duration (Days)")
plt.ylabel("Number of Documents")
plt.tight_layout()
plt.show()
```
